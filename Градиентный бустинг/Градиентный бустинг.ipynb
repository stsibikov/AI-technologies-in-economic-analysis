```python
X = df.drop(columns='Answer')
y = df.Answer
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 888)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

clf = GradientBoostingClassifier().fit(X_train, y_train)
clf
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

y_pred = clf.predict(X_test)

y_true = y_test

accuracy = accuracy_score(y_true, y_pred)
print(f'Accuracy: {accuracy:.4f}')

precision = precision_score(y_true, y_pred)
print(f'Precision: {precision:.4f}')

recall = recall_score(y_true, y_pred)
print(f'Recall: {recall:.4f}')

f1 = f1_score(y_true, y_pred)
print(f'F1 Score: {f1:.4f}')

conf_matrix = confusion_matrix(y_true, y_pred)
print('Confusion Matrix:')
print(conf_matrix)

class_report = classification_report(y_true, y_pred)
print('Classification Report:')
print(class_report)
```

## Задание 2. Обучите модель случайного леса и сравните результаты с градиентным бустингом

```python
y_pred = randomforest.predict(X_test)
y_true = y_test

accuracy = accuracy_score(y_true, y_pred)
print(f'Accuracy: {accuracy:.4f}')

precision = precision_score(y_true, y_pred)
print(f'Precision: {precision:.4f}')

recall = recall_score(y_true, y_pred)
print(f'Recall: {recall:.4f}')

f1 = f1_score(y_true, y_pred)
print(f'F1 Score: {f1:.4f}')

conf_matrix = confusion_matrix(y_true, y_pred)
print('Confusion Matrix:')
print(conf_matrix)

class_report = classification_report(y_true, y_pred)
print('Classification Report:')
print(class_report)
```

```python
pd.Series(index=randomforest.feature_names_in_, data=randomforest.feature_importances_).sort_values(ascending = False)
```

**Удаление 5 переменных из датасета для новых моделей**  
В этом задании необходимо удалить переменные исходя только из здравого смысла.
* `passport` - можно предположить, что наличие загранпаспорта позволяет сотрудникам, например, ездить на иностранные конференции для обмена идеями, что помогло бы компании быть более инновационной. Однако, анализ этого датасета показал, что корреляция между `passport` и `Answer` негативная. Так, такая корреляция не имеет никакого смысла, поэтому мы ее убираем
* `wifi` - наличие открытой сети является скорее решением ИТ поддержки, и, соответственно, может служить индикатором деятельности только службы ИТ поддержки
* `green` - озеленненность офиса (кол-во растений) хоть и решается сотрудниками компании, но он может служить только индикатором любви сотрудников к растениям
* `mortgage` - наличие ипотеки у сотрудников не влияет на инновационность компании
* `papers` - научные статьи содержат теоретическую основу, а потому редко приводят к каким-либо инновациям. Также нужно иметь ввиду, что существуют разные типы инноваций - например, маркетинговая или организационная инновация - к которой научная статья явно бы не привела. Из-за крайне слабой связи переменная убирается

## Задание 3. Обучите модель градиентного бустинга на новых переменных

```python
df_2 = df.drop(columns=['passport', 'wifi', 'green', 'mortgage', 'papers'])

X = df_2.drop(columns='Answer')
y = df_2['Answer']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1234)
```

```python
clf = GradientBoostingClassifier(max_depth=3).fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_true = y_test

accuracy = accuracy_score(y_true, y_pred)
print(f'Accuracy: {accuracy:.4f}')

precision = precision_score(y_true, y_pred)
print(f'Precision: {precision:.4f}')

recall = recall_score(y_true, y_pred)
print(f'Recall: {recall:.4f}')

f1 = f1_score(y_true, y_pred)
print(f'F1 Score: {f1:.4f}')

conf_matrix = confusion_matrix(y_true, y_pred)
print('Confusion Matrix:')
print(conf_matrix)

class_report = classification_report(y_true, y_pred)
print('Classification Report:')
print(class_report)
```

```python
pd.Series(index = clf.feature_names_in_, data = clf.feature_importances_).sort_values(ascending = False)
```

## Задание 4. Обучите модель случайного леса на новых переменных и сравните результаты с градиентным бустингом

```python
randomforest = ensemble.RandomForestClassifier(max_depth=3)
randomforest.fit(X_train, y_train)
```

```python
y_pred = randomforest.predict(X_test)
y_true = y_test

accuracy = accuracy_score(y_true, y_pred)
print(f'Accuracy: {accuracy:.4f}')

precision = precision_score(y_true, y_pred)
print(f'Precision: {precision:.4f}')

recall = recall_score(y_true, y_pred)
print(f'Recall: {recall:.4f}')

f1 = f1_score(y_true, y_pred)
print(f'F1 Score: {f1:.4f}')

conf_matrix = confusion_matrix(y_true, y_pred)
print('Confusion Matrix:')
print(conf_matrix)

class_report = classification_report(y_true, y_pred)
print('Classification Report:')
print(class_report)
```

```python
pd.Series(index = randomforest.feature_names_in_, data = randomforest.feature_importances_).sort_values(ascending = False)
```

**Сравнение моделей, обученных на новых переменных**  
На новых переменных модель градиентного бустинга показала себя лучше модели случайного леса, что видно по accuracy (93 против 88), а также по confusion matrix. Модель градиентного бустинга сделала меньше ошибок, особенно второго типа (False Negative).

В качестве главной метрики сравнения моделей принимаем F-1 score, так как этот показатель является средним гармоническим между precision и recall. Данная метрика лучше у модели градиентного бустинга (93 против 86) и мы принимаем ее как оптимальную.
